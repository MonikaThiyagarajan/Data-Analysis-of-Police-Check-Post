# Data Analysis on Police Dataset

The data from a Police check post is given.

##### Import Libraries

```python
import pandas as pd
```

```python
data = pd.read_csv("Police Data.csv")
```

```python
data
```

# Remove the column that only contains missing values

```python
data.isnull().sum()
```

```python
data.drop(['country_name'],inplace = True,axis=1)
```

```python
data.head()
```

# Based on Filtering + Value Counts

# For speeding , were men or women stopped more often?

```python
data[data.violation == 'Speeding'].driver_gender.value_counts()
```

For speeding nearly 25517 men were stopped in this dataset.

# Groupby

# Who gets searched during a stop? Does gender affect?

```python
data.head(2)
```

```python
data.groupby('driver_gender').search_conducted.sum()
```

Men gets searched often during a stop in this data.

```python
data.search_conducted.value_counts()
```

In 2479 searches conducted by police, men was searched 2113.

# Mapping + Data-type casting

# What is the mean of Stop duration?

```python
data.stop_duration.value_counts()
```

```python
#type casting
data['stop_duration']=data['stop_duration'].map({'0-15 Min':7.5,'16-30 Min':24,'30+ Min':45})
```

```python
data.head()
```

```python
#mean
data['stop_duration'].mean()
```
